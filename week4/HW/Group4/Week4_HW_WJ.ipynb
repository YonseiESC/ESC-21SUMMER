### HW2 : PCA from scratch  
\- iris 데이터에 numpy만을 사용해서 PCA를 해봅시다!  
\- We can find PCA makes visualization easier!

```python
import numpy as np 
import pandas as pd

from sklearn.preprocessing import StandardScaler

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/uiuc-cse/data-fa14/gh-pages/data/iris.csv')
df.head()
```

```python
X = df.iloc[:,:-1]    # 마지막 열만 자르기
label = df.iloc[:,-1]
X.head()
```

#### Using Covariance Matrix

```python
# Step 1. Center Data
X_scaled = StandardScaler().fit_transform(X)   # Scaling X
X_scaled[:5]
```

```python
# Step 2. Compute Covariance Matrix
cov_matrix = (X_scaled.T @ X_scaled) / (X_scaled.shape[0]-1)
# X.T@X / (n-1) ... X가 n > p 이기 때문에 X.T@X. p=4이므로 cov_matrix는 4x4 matrix.
cov_matrix
```

```python
# Step 3. Eigenvalue decomposition
eigvals, eigvecs = np.linalg.eig(cov_matrix)  # cov_matrix의 eigenvalues.
eigvals
```

```python
# Ratio of explained variance per PC
explained_variances = []
for i in range(len(eigvals)):
    explained_variances.append(eigvals[i] / np.sum(eigvals))
    
 # PC의 variance = eigenvalue이기 때문에, eigenvalue로 variance의 설명력을 구한다!

print(np.sum(explained_variances), '\n', explained_variances)
```

첫 번째, 두 번째 PC가 이미 variance의 95% 이상을 설명함을 확인할 수 있다!

```python
# Visualization (Embedding)
pc1 = X_scaled @ eigvecs[:, 0]     # X_scaled에 eigenvector를 곱해서 PC를 구한다!
pc2 = X_scaled @ eigvecs[:, 1]
res = pd.DataFrame(pc1, columns=['PC1'])
res['PC2'] = pc2
res['label'] = label
res.head()
```

```python
# Projection on 1-dim subspace
plt.figure(figsize=(20, 10))
sns.scatterplot(res['PC1'], [0] * len(res), hue=res['label'], s=200)
```

```python
# Projection on 2-dim subspace
plt.figure(figsize=(20, 10))
sns.scatterplot(res['PC1'], res['PC2'], hue=res['label'], s=100)
```
