```python
# Data Import
import ssl
import pandas as pd
ssl._create_default_https_context = ssl._create_unverified_context #Github     .       !
data = pd.read_csv('https://github.com/YonseiESC/ESC-21SUMMER/blob/main/week1/HW/week1_data.csv?raw=True')
y = data['mpg']
x = data.drop(['mpg'],axis=1)
```

```python
x['horsepower'].unique()
```

```python
idx = x['horsepower']!='?'
x = x[idx]
y = y[idx]
x = x.astype({'horsepower':'float'})
```

```python
import numpy as np
# numpy  .
def MyOwnRegression(x,y):
    betahat = np.linalg.inv(x.T@x)@x.T@y
    yhat = x@betahat
    return(betahat, yhat)
    
```

```python
x = np.array(x)
y = np.array(y)
MyOwnRegression(x, y)
```
