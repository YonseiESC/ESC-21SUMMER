```python
import ssl
import pandas as pd
import numpy as np
ssl._create_default_https_context = ssl._create_unverified_context
data = pd.read_csv('https://github.com/YonseiESC/ESC-21SUMMER/blob/main/week1/HW/week1_data.csv?raw=True')
```

```python
y = data['mpg']
x = data.drop(['mpg'],axis=1)
```

```python
x.info()
```

```python
x["horsepower"].unique()
```

```python
x["horsepower"] = [0 if i == '?' else i for i in x["horsepower"]]
x["horsepower"] = x["horsepower"].astype("float")
```

```python
y = np.array(y)
x = np.array(x)
```

```python
def YourOwnRegression(x,y):
    x_t = np.transpose(x)
    beta_hat = np.linalg.inv(x_t @ x) @ x_t @ y
    return [beta_hat, x @ beta_hat]
```

```python
YourOwnRegression(x,y)
```

```python
# check
[beta_hat, y_hat] = YourOwnRegression(x,y)
print(x[0] @ beta_hat)
```
