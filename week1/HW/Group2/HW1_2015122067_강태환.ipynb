```python
# Data Import
import ssl
import pandas as pd
ssl._create_default_https_context = ssl._create_unverified_context #Github     .       !
data = pd.read_csv('https://github.com/YonseiESC/ESC-21SUMMER/blob/main/week1/HW/week1_data.csv?raw=True')
y = data['mpg']
x = data.drop(['mpg'],axis=1)
```

```python
data2 = data.astype({'horsepower': 'float'})
y = data2['mpg']
x = data2.drop(['mpg'],axis=1)
```

```python
# numpy  .
def Beta(X,Y):
    # beta y    .
    beta= np.dot((np.linalg.inv(np.dot(X.T,X))), np.dot(X.T,Y))
    y_hat=np.dot(X,beta)
    return beta,y_hat

#  
```

```python
beta,y_hat =Beta(x,y)
beta
```
