```python
data2 = data.astype({'horsepower': 'float'})
y = data2['mpg']
x = data2.drop(['mpg'],axis=1)
```

```python
# numpy 모듈만을 이용해주세요.
def Beta(X,Y):
    # beta와 y의 추정값을 반환하는 함수를 만들어주세요.
    beta= np.dot((np.linalg.inv(np.dot(X.T,X))), np.dot(X.T,Y))
    y_hat=np.dot(X,beta)
    return beta,y_hat

# 결과물 반환
```

```python
beta,y_hat =Beta(x,y)
beta
```
