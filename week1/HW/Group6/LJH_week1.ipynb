\hat{y_i} = x_i\hat{\beta} \\
\hat{\beta} = (\Sigma_{j=1}^{n}x_jy_j)/(\Sigma_{j=1}^{n}x_j^2) \\
\hat{y_i} = \Sigma_{j=1}^{n}a_jy_j
$
  
$
\begin{align}
\hat{y_i} &= x_i (\Sigma_{j=1}^{n}x_jy_j)/(\Sigma_{j=1}^{n}x_j^2) \\
& = (\Sigma_{j=1}^{n}x_ix_jy_j)/(\Sigma_{j=1}^{n}x_j^2) \\
& = \Sigma_{j=1}^{n}a_jy_j
\end{align}
$
  
$
a_j = \frac{\Sigma_{j=1}^{n}x_ix_j}{\Sigma_{j=1}^{n}x_j^2}

  x_1\cdot q_1 & x_2\cdot q_1 & ... & x_p\cdot q_1 \\

  0 & 0 & ... & x_p\cdot q_p
}\right] \\
\end{align}
$

\begin{align}

&= ((QR)^{T}QR)^{-1}(QR)^{T}Y \\

&= (R^TR)^{-1}R^{T}Q^{T}Y \:\: \because QQ^{T}=Q^{T}Q=I_n \\
&= R^{-1}(R^{T})^{-1}R^{T}Q^{T}Y \\
&= R^{-1}Q^{T}Y
\end{align}

```python
import numpy as np
import ssl
import pandas as pd
ssl._create_default_https_context = ssl._create_unverified_context #Github에서 데이터를 바로 불러오도록 하는 세팅입니다. 해당 코드 무시하고 데이터 받아서 쓰셔도 됩니다!
data = pd.read_csv('https://github.com/YonseiESC/ESC-21SUMMER/blob/main/week1/HW/week1_data.csv?raw=True')
data = data[data['horsepower']!='?']
x = data[['cylinders','displacement','weight','acceleration','year']].to_numpy()
y = data['mpg'].to_numpy()
```

```python
def MyRegression(Y, Z, get):
    betahat = np.linalg.inv(Z.transpose().dot(Z)).dot(Z.transpose()).dot(Y)
    Yhat = Z.dot(betahat) 
    
    if get == 'beta': # beta_hat matrix
        return np.round(pd.DataFrame(betahat),4)
    
    elif get == 'Y': # Y_hat matrix
        return np.round(Yhat,4)
```

```python
MyRegression(y, x, 'beta')
```

```python
MyRegression(y, x, 'Y')
```
