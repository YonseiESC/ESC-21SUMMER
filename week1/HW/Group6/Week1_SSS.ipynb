```python
# Data Import
import ssl
import pandas as pd
import numpy as np
ssl._create_default_https_context = ssl._create_unverified_context 
#Github     .       !
data = pd.read_csv('https://github.com/YonseiESC/ESC-21SUMMER/blob/main/week1/HW/week1_data.csv?raw=True')
y = data['mpg']
x = data.drop(['mpg'],axis=1)
```

```python
row = x[x['horsepower']=='?'].index
x = x[row]
y = y[row]
x['horsepower']=x['horsepower'].astype(float)
```

```python
# numpy  .
def YourOwnRegression(x,y):
    beta = np.dot((np.linalg.inv(np.dot(x.T,x))), np.dot(x.T,y))
    y_hat = np.dot(x, beta)
    return beta, y_hat

#  
YourOwnRegression(x,y)
```
