```python
def YourOwnRegression(x,y):
    beta = np.dot(np.dot(np.linalg.inv(np.dot(x.T, x)), x.T), y) 
    y_hat = np.dot(x, beta)
    return beta, y_hat

# ê²°ê³¼ë¬¼ ë°˜í™˜
YourOwnRegression(x,y)
```
