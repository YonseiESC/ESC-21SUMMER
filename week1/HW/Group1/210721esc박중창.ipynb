```python
# Data Import
import ssl
import pandas as pd
ssl._create_default_https_context = ssl._create_unverified_context 
data = pd.read_csv('https://github.com/YonseiESC/ESC-21SUMMER/blob/main/week1/HW/week1_data.csv?raw=True')
```

```python
data.info()
```

```python
import numpy as np
```

```python
# convert type(object -> float)
idx = np.where(data['horsepower'] == '?') #thanks to. Seoyeon Ahn
data = data.drop(idx[0])
data = data.astype({'horsepower': 'float'})
```

```python
data.info()
```

```python
# set x, y
X = data.drop(['mpg'],axis=1)
y = data['mpg']
X = X.to_numpy()
y = y.to_numpy()
```

```python
def Regression_Std(x,y):
    beta_h = np.linalg.inv(x.T@x)@x.T@y 
    y_h = x@beta_h
    return beta_h, y_h

def Regression_Gram(x,y):
    Q, R = np.linalg.qr(x)
    beta_h = np.linalg.inv(R)@Q.T@y
    y_h = x@beta_h
    return beta_h, y_h

b_s,y_s =Regression_Std(X,y)
b_g,y_g =Regression_Gram(X,y)
```

```python
print('b_s :',b_s)
print('b_g :',b_g)
```
