```python
import pandas as pd
import numpy as np

data = pd.read_csv('https://github.com/YonseiESC/ESC-21SUMMER/blob/main/week1/HW/week1_data.csv?raw=True')
y = data['mpg']
x = data.drop(['mpg'],axis=1)
```

```python
x.info()
```

```python
x
```

```python
remove_ind = x[~x['horsepower'].str.isnumeric()].index
x = x.drop(remove_ind).reset_index(drop=True)
y = y.drop(remove_ind).reset_index(drop=True)
x['horsepower'] = x['horsepower'].astype('float')
```

```python
def LinearRegression(X, y):
    X_t = np.transpose(X)
    beta = np.linalg.inv(X_t@X)@X_t@y
    y_hat = X@beta
    return y_hat
```

```python
x_train = x.values
y_train = y.values
output = LinearRegression(x_train,y_train)
```
